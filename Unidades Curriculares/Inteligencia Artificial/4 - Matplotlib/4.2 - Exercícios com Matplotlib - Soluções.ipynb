# Exercícios

Siga as instruções para recriar os gráficos usando esses dados:

## Dados

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
x
```

**Importe matplotlib.pyplot como plt e defina %matplotlib inline se estiver usando o jupyter notebook. Qual comando você utiliza se não estiver usando o jupyter notebook?**

```python
import matplotlib.pyplot as plt
%matplotlib inline
# plt.show() for non-notebook users
```

## Exercício 1

**Siga estes passos:**
* **Crie um objeto figure chamado fig usando plt.figure()**
* **Use add_axes para adicionar um eixo à tela da figura em [0,0,1,1]. Chame esse novo eixo de ax.**
* **Plote (x,y) nesses eixos e defina os rótulos e títulos para corresponder ao gráfico abaixo:**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

## Exercício 2

**Crie um objeto figura e coloque dois eixos nele, ax1 e ax2. Localizado em [0,0,1,1] e [0.2,0.5,.2,.2] respectivamente.**

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
```

**Agora plote (x,y) em ambos os eixos e chame seu objeto figure para mostrá-lo.**

```python
ax1.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')


ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')

fig # Mostra objeto figure
```

## Exercício 3

**Crie o gráfico abaixo adicionando dois eixos a um objeto de figura em [0,0,1,1] e [0.2,0.5,.4,.4]**

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

**Agora utilize x, y e z para recriar o gráfico abaixo. Observe os limites x e y no gráfico inserido:**

```python
ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')


ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

fig
```

## Exercício 4

**Use plt.subplots(nrows=1, ncols=2) para criar o gráfico abaixo.**

```python
# Canvas vazio de 1 por 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=2)
```

**Agora plote (x,y) e (x,z) nos eixos. Brinque com a largura de linha e o estilo**

```python
axes[0].plot(x,y,color="blue", lw=3, ls='--')
axes[1].plot(x,z,color="red", lw=3, ls='-')
fig
```

**Veja se você pode redimensionar o gráfico adicionando o argumento figsize() em plt.subplots() em seu código anterior.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))

axes[0].plot(x,y,color="blue", lw=5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z,color="red", lw=3, ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```

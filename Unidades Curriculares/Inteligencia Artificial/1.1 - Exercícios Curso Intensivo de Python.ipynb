**Crie uma função que pegue o domínio do site de email de uma string na forma:**

    user@domain.com
    
**Por exemplo, passar "usuário@domínio.com" retornaria: domínio.com**

```python
def get_domain(email):
  """
  Extrai o domínio do email de uma string.

  Args:
    email: A string contendo o endereço de email.

  Returns:
    A string contendo o domínio do email.
  """
  return email.split('@')[1]

# Example usage:
email_address = "usuario@dominio.com"
domain = get_domain(email_address)
print(domain)

```

**Crie uma função básica que retorne True se a palavra 'dog' estiver contida na string de entrada. Não se preocupe com casos extremos, como uma pontuação sendo anexada à palavra dog, mas leve em consideração a capitalização.**

```python
def return_dog(string):
  """
  Retorna True se a palavra 'dog' estiver contida na string de entrada.

  Args:
    string: A string de entrada.

  Returns:
    True se a palavra 'dog' estiver contida na string, False caso contrário.
  """
  if  'dog' in string:
    return True
  else:
    return False

print(return_dog('dog'))
```

**Crie uma função que conte o número de vezes que a palavra "dog" ocorre em uma string. Novamente ignore os casos extremos.**

```python
def count_dog(text):
  """
  Conta o número de vezes que a palavra "dog" ocorre em uma string.

  Args:
    text: A string de entrada.

  Returns:
    O número de ocorrências da palavra "dog".
  """
  words = text.lower().split()  # Convert to lowercase and split into words
  count = 0
  for word in words:
    if 'dog' in word:  # Check if 'dog' is a substring (handles simple cases)
      count += 1
  return count

# Example usage:
print(count_dog("Dog is a good dog."))
print(count_dog("All cats are not dogs."))
print(count_dog("Dog dog dog."))
```

**Use expressões lambda e a função filter() para filtrar palavras de uma lista que não comecem com a letra 's'. Por exemplo:**

    seq = ['soup','dog','salad','cat','great']

**deve ser filtrado para:**

    ['soup','salad']

```python
seq = [ 'soup','dog','salad','cat','great']

# Use filter and a lambda function to keep words that start with 's'
filtered_seq = list(filter(lambda word: word.startswith('s'), seq))

print(filtered_seq)
#
```

** Escreva um script que leia um número digitado pelo usuário e entregue uma lista de todos os números calculados para encontrar o fatorial do que foi digitado, por exemplo, usuário digita 5, ele deve apresentar [20, 60, 120, 120], este valores devem estar armazenados em uma lista

```python
def factorial_list(n):
  """
  Calcula o fatorial de um número e retorna uma lista com os resultados intermediários.

  Args:
    n: O número para calcular o fatorial.

  Returns:
    Uma lista com os resultados intermediários do cálculo do fatorial.
  """
  if n < 0:
    return "Fatorial não definido para números negativos"
  elif n == 0:
    return [1]
  else:
    result = 1
    factorial_steps = []
    for i in range(1, n + 1):
      result *= i
      factorial_steps.append(result)
    return factorial_steps

# Get input from the user
try:
  num = int(input("Digite um número inteiro não negativo: "))
  print(factorial_list(num))
except ValueError:
  print("Entrada inválida. Por favor, digite um número inteiro.")
```

###Problema final
**Você está dirigindo um pouco rápido demais e um policial o para. Escreva uma função para retornar um dos 3 resultados possíveis: "No ticket", "Small ticket" ou "Big Ticket".
  Se a sua velocidade for 60 ou menos, o resultado é "Sem Bilhete". Se a velocidade estiver entre 61 e 80 inclusive, o resultado é "Small Ticket". Se a velocidade for 81 ou mais, o resultado é "Big Ticket". A menos que seja seu aniversário (codificado como um valor booleano nos parâmetros da função) -- em seu aniversário, sua velocidade pode ser 5 unidades maior em todos casos. **

```python
def caught_speeding(speed, is_birthday):
  """
  Determina o tipo de multa por excesso de velocidade.

  Args:
    speed: A velocidade do veículo.
    is_birthday: Um valor booleano indicando se é o aniversário do motorista.

  Returns:
    Uma string indicando o tipo de multa ("No ticket", "Small ticket", ou "Big Ticket").
  """
  speed_limit = 60
  small_ticket_limit = 80

  if is_birthday:
    speed_limit += 5
    small_ticket_limit += 5

  if speed <= speed_limit:
    return "No ticket"
  elif speed > speed_limit and speed <= small_ticket_limit:
    return "Small ticket"
  else:
    return "Big Ticket"

# Example usage:
print(caught_speeding(60, False))
print(caught_speeding(65, False))
print(caught_speeding(85, False))
print(caught_speeding(65, True))
print(caught_speeding(70, True))
print(caught_speeding(90, True))
```

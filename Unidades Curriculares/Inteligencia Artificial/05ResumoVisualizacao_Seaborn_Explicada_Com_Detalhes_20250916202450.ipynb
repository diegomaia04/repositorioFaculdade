# Visualização de Dados com Seaborn – Aula Completa

Este notebook contém explicações detalhadas e comandos para criar diferentes tipos de gráficos com a biblioteca Seaborn. Ideal para quem está começando na área de visualização de dados.

```python
# Importa a biblioteca Seaborn para visualizações estatísticas e Matplotlib para exibição dos gráficos.
# Também carrega um dataset de exemplo chamado 'tips', que contém dados de gorjetas em um restaurante.
# Por fim, exibe as primeiras linhas do dataset com o método head().
# Importação das bibliotecas

import seaborn as sns

import matplotlib.pyplot as plt



# Carregamento de um dataset exemplo

tips = sns.load_dataset('tips')

tips.head()
```

## 1. Gráficos de Distribuição
**Objetivo:** Mostrar como os dados numéricos estão distribuídos.

### Histogramas com `histplot`

```python
# Cria um histograma da coluna 'total_bill' para mostrar a frequência das contas totais.
# Esse tipo de gráfico ajuda a entender como os valores estão distribuídos ao longo do eixo X.
sns.histplot(tips['total_bill'])

plt.title('Histograma de Total da Conta')

plt.show()
```

### Linha de densidade com `kdeplot`

```python
# Cria uma curva de densidade com base na coluna 'total_bill', usando a técnica KDE (Kernel Density Estimation).
# É uma maneira suave de visualizar a distribuição de uma variável contínua.
sns.kdeplot(tips['total_bill'])

plt.title('Estimativa de Densidade de Total da Conta')

plt.show()
```

### Gráfico combinado com `displot`

```python
# Combina histograma com curva de densidade para fornecer uma visão mais completa da distribuição da variável.
sns.displot(tips['total_bill'], kde=True)

plt.title('Histograma + KDE de Total da Conta')

plt.show()
```

## 2. Gráficos Categóricos
**Objetivo:** Comparar variáveis numéricas com variáveis categóricas.

### Boxplot

```python
# Cria um boxplot da variável 'total_bill' separada por 'day'.
# Boxplots mostram mediana, quartis, outliers e distribuição geral de forma compacta.
sns.boxplot(x='day', y='total_bill', data=tips)

plt.title('Boxplot por dia da semana')

plt.show()
```

### Violinplot

```python
# Cria um violinplot, que é similar ao boxplot, mas mostra também a densidade da distribuição.
# É útil para comparar distribuições entre categorias.
sns.violinplot(x='day', y='total_bill', data=tips)

plt.title('Violinplot por dia da semana')

plt.show()
```

### Barplot e Countplot

```python
# Cria um barplot, que mostra a média de 'total_bill' por dia.
# Depois, um countplot, que mostra a contagem de observações por categoria de 'day'.
sns.barplot(x='day', y='total_bill', data=tips)

plt.title('Barplot por dia da semana')

plt.show()



sns.countplot(x='day', data=tips)

plt.title('Countplot de dias da semana')

plt.show()
```

## 3. Matrix Plots
**Objetivo:** Visualizar correlações ou agrupamentos em formato de matriz.

### Heatmap de Correlação

```python
# Calcula a matriz de correlação entre variáveis numéricas e a plota com um heatmap.
# O argumento 'annot=True' mostra os valores dentro da matriz.
# O parâmetro 'cmap' define o esquema de cores.
corr = tips.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title('Mapa de Calor de Correlação')

plt.show()
```

## 4. Grids
**Objetivo:** Comparar múltiplos gráficos lado a lado com FacetGrid e Pairplot.

### Pairplot

```python
# Cria um pairplot, que mostra todas as combinações possíveis de scatter plots entre variáveis numéricas.
# Usa a cor para separar os pontos com base na variável 'sex'.
sns.pairplot(tips, hue='sex')

plt.suptitle('Pairplot do dataset Tips', y=1.02)

plt.show()
```

### FacetGrid

```python
# Cria um grid de histogramas com a função FacetGrid, onde cada gráfico representa um valor distinto de 'sex'.
# É útil para comparar distribuições entre grupos.
g = sns.FacetGrid(tips, col='sex')

g.map(sns.histplot, 'total_bill')

plt.suptitle('FacetGrid por sexo', y=1.05)

plt.show()
```

## 5. Regressão
**Objetivo:** Visualizar relações lineares entre duas variáveis numéricas.

### regplot e lmplot

```python
# Plota uma linha de regressão simples com base nas variáveis 'total_bill' e 'tip'.
# O regplot mostra os dados e ajusta uma linha de regressão linear simples.
# O lmplot permite separação por categoria (hue) e é mais flexível para painéis.
sns.regplot(x='total_bill', y='tip', data=tips)

plt.title('Regressão Linear com regplot')

plt.show()



sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')

plt.title('Regressão Linear com lmplot')

plt.show()
```

## 6. Estilo e Cores
**Objetivo:** Personalizar o visual dos gráficos.

### Estilo, Contexto e Paleta de Cores

```python
# Personaliza os gráficos com estilo ('whitegrid'), contexto ('talk', que aumenta o tamanho das fontes para apresentações)
# e uma paleta de cores pastel. Depois, exibe um boxplot estilizado.
sns.set_style('whitegrid')

sns.set_context('talk')

sns.set_palette('pastel')



sns.boxplot(x='day', y='total_bill', data=tips)

plt.title('Boxplot Estilizado')

plt.show()
```
